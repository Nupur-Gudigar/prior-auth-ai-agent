```python
# Cell 3 вЂ” Test AWS Bedrock + Llama 3.1 connection
import boto3
import json

# Your AWS credentials
AWS_ACCESS_KEY = dbutils.secrets.get(scope="aws-credentials", key="aws-access-key")
AWS_SECRET_KEY = dbutils.secrets.get(scope="aws-credentials", key="aws-secret-key")
AWS_REGION = "us-east-2"  # Ohio вЂ” where your account is

# Connect to Bedrock
bedrock = boto3.client(
    service_name="bedrock-runtime",
    region_name=AWS_REGION,
    aws_access_key_id=AWS_ACCESS_KEY,
    aws_secret_access_key=AWS_SECRET_KEY
)

# Send a test message to Llama 3.1 8B
body = json.dumps({
    "prompt": "<|begin_of_text|><|start_header_id|>user<|end_header_id|>Say hello and confirm you are ready to help with prior authorization decisions.<|eot_id|><|start_header_id|>assistant<|end_header_id|>",
    "max_gen_len": 200,
    "temperature": 0.1
})

response = bedrock.invoke_model(
    modelId="us.meta.llama3-1-8b-instruct-v1:0",
    body=body
)

result = json.loads(response["body"].read())
print(result["generation"])
```
